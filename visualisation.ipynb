```python
def param(Type,Zone):
    a_gr,a_g,T_b,T_c,T_d,SS,gamma,mu= 0,0,0,0,0,0,0,0
    if Type=='A' and Zone ==5:
        a_g=1.6
        T_b=0.15
        T_c=0.4
        T_d=2
        SS=1
        gamma=1.4
        mu=0.8
    if Type=='B'and Zone==5:
        a_g=1.6
        T_b=0.15
        T_c=0.5
        T_d=2
        SS=1.2
        gamma=1
        mu=0.6
    if Type=='A' and 1<=Zone<=4:
        if Zone==1:
            a_gr=0.4
        if Zone==2:
            a_gr=0.7
        if Zone==3:
            a_gr=1.1
        if Zone==4:
            a_gr=2
        gamma=1
        a_g=gamma*a_gr
        T_b=0.03
        T_c=0.2
        T_d=2.5
        SS=1
        gamma=1
        mu=0.6
    if Type=='B'and  1<=Zone<=4:
        if Zone==1:
            a_gr=0.4
        if Zone==2:
            a_gr=0.7
        if Zone==3:
            a_gr=1.1
        if Zone==4:
            a_gr=2
        gamma=1
        a_g=a_gr*gamma
        T_b=0.03
        T_c=0.25
        T_d=2.5
        SS=1.35
        
        mu=0.6
    return a_gr,a_g,T_b,T_c,T_d,SS,gamma,mu


    

def S(T):
    Se=0
    if (0<=T and T<=T_b):
        Se=a_g*SS*(1+(T/T_b)*((2.5*mu)-1))
    if (T_b<=T and T<=T_c):
        Se=a_g*SS*(2.5)*mu
    if (T_c<=T and T<=T_d):
        Se=a_g*SS*(T_c/T)*2.5*mu
    if (T_d<=T and T<=4):
        Se=a_g*SS*(T_c*T_d/(T**2))*(2.5)*mu
    return Se
    
```

# Spectres de calcul pour l’analyse élastique avec 𝒒 ≥ 𝟏


    
Calcul de q

q=q_0*kw

On modélise le système par un ensemble de cadres (frame system) selon l'axe x, on prend alors q0=3*alpha0/alpha1 pour une DCM et q0=4.5*alpha0/alpha1 pour une DCH, la valeur de alpha0/alpha1 vaut 1.1 pour un multi-bay frame system on néglige également la torsion selon x, ce qui nous donne kw=1 ce qui nous donne une valeur de q(DCM)=3.3 et q(DCH)=4.4



```python
beta=0.2
q=1.5


def Sd(T):
    Se=0
    if (0<=T and T<=T_b):
        Se=a_g*SS*((2/3)+(T/T_b)*((2.5/q)*0.7-(2/3)))
    if (T_b<=T and T<=T_c):
        Se=a_g*SS*(2.5)*mu/q
    if (T_c<=T and T<=T_d):
        Se=max((a_g*SS*(T_c/T)*2.5*mu)/q,beta*a_g)
    if (T_d<=T and T<=4):
        Se=max((a_g*SS*(T_c*T_d/(T**2))*(2.5)*mu)/q,beta*a_g)
    return Se

a_gr,a_g,T_b,T_c,T_d,SS,gamma,mu=param('B',4)
T=[0.1*i for i in range(40)]
q=3.3
Sld=[Sd(T[i]) for i in range(40)]
q=4.4
Sld2=[Sd(T[i]) for i in range(40)]

plt.plot(T, Sl)
plt.xlabel('T')
plt.ylabel('S')
plt.show()
```

```python
plt.plot(T, Sl, color='r', label='q(DCH)=4.4')
plt.plot(T, Sld, color='g', label='q(DCM)=3.3')
plt.plot(T, Sld2, color='b', label='q=1')
  
# Naming the x-axis, y-axis and the whole graph
plt.xlabel("T")
plt.ylabel("S")
plt.title("SEISME")
  
# Adding legend, which helps us recognize the curve according to it's color
plt.legend()
  
# To load the display window
plt.show()
```
